```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Input scrayping data as dataframe including myself 
DIR = '/content/drive/My Drive/Colab Notebooks/rottenntomatos/input/'
df = pd.read_csv(DIR + 'df_rotten-tomatoes_hiro.csv',sep=',')
df.head()
```

```
# Extract titles more than 10 reviews
vc_title = df['Title'].value_counts()
vc_title_n10_more = vc_title[vc_title>10]

# Make new dataframe 
df2 = df[df['Title'].isin(vc_title_n10_more.index)]

# Check numbers of titles/reviewers
print('number of titles:{}'.format(len(df2['Title'].unique())))
print('number of Reviewer:{}'.format(len(df2['Reviewer'].unique())))
```

```
# Cross tabulation of Titles and Reviewer
crosstab = pd.crosstab(df2['Title'],df2['Reviewer'],values=df2['Rate'],aggfunc='mean')
crosstab.head()
```

```
# Replace NaN to mean values at each titles
crosstab2 = crosstab.T
crosstab3 = crosstab2.fillna(crosstab2.mean())
crosstab3.head()
```

```
# Clustering with kmeans(I set 10 clusters. There is no special reason)
from sklearn.cluster import KMeans
N_CLUSTERS = 20
FEATURES = crosstab3.columns

crosstab3_T=crosstab3.T
cls = KMeans(n_clusters=N_CLUSTERS)
pred = cls.fit_predict(crosstab3)
centers = cls.cluster_centers_
pd.options.display.precision = 2
pd.DataFrame(centers,columns=FEATURES)
```

```
# Add clustered ID to the DataFfame
crosstab3['cluster_id']=pred
crosstab3.head()
```

```
# Distribution of each cluster
crosstab3['cluster_id'].value_counts()
```

```
# Check my cluster ID
titles = pd.Series(crosstab3.index)
clustered_id = pd.Series(pred)
df_clustered = pd.concat([titles, clustered_id], axis=1)
df_clustered[df_clustered['Reviewer'] == 'MYNAME']
```
